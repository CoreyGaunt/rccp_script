```python
import pandas as pd
import numpy as np
import os
import csv
from itertools import chain
```

```python
file = "test_data_KINDRED_2.csv"

csv = pd.read_csv(file,encoding="ISO-8859-1")
df = pd.DataFrame(csv)

csv.head()
```

```python
df1 = df.dropna()
df1
```

```python
# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# calculate lengths of splits
lens = df1['Flat RCCP'].str.split(',').map(len)
```

```python
# creating a new dataframe with the flat images separated out of each other

flat = pd.DataFrame({
    'Manufacturer':np.repeat(df1['Manufacturer'], lens),
    'Series':np.repeat(df1['Series'], lens),
    'Color':np.repeat(df1['Color'], lens),
    'Colors #':np.repeat(df1['Colors #'], lens),
    'Pattern Reference\n(URL Links)':np.repeat(df1['Pattern Reference\n(URL Links)'], lens),
    'Flat RCCP':chainer(df1['Flat RCCP']),
    'Presence in DMP':np.repeat(df1['Presence in DMP'], lens),
    'Material Creation Status':np.repeat(df1['Material Creation Status'], lens)
})
```

```python
flat
```

```python
# calculate lengths of splits
lens2 = df1['Texture RCCP'].str.split(',').map(len)

# creating a new dataframe with the texture images separated out of each other

texture = pd.DataFrame({
    'Manufacturer':np.repeat(df1['Manufacturer'], lens2),
    'Series':np.repeat(df1['Series'], lens2),
    'Color':np.repeat(df1['Color'], lens2),
    'Colors #':np.repeat(df1['Colors #'], lens2),
    'Pattern Reference\n(URL Links)':np.repeat(df1['Pattern Reference\n(URL Links)'], lens2),
    'Texture RCCP':chainer(df1['Texture RCCP']),
    'Presence in DMP':np.repeat(df1['Presence in DMP'], lens2),
    'Material Creation Status':np.repeat(df1['Material Creation Status'], lens2)
})
```

```python
texture
```

```python
texture['Texture RCCP'] = texture['Texture RCCP'].str.replace('Use ','')
```

```python
texture.head(20)
```

```python
merged = pd.merge(flat,texture,on='Color', how='inner')
```

```python
merged.head()
```

```python
merged2 = merged[['Manufacturer_x','Series_x','Color','Colors #_x','Flat RCCP','Texture RCCP']]
clean_merge = merged2.rename(columns={'Manufacturer_x':'Manufacturer',
                       'Series_x':'Series',
                       'Colors #_x':'Colors #'
                       })
```
